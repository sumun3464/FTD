```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

#1. KNN
#iris 데이터 로드
iris = load_iris()
#knn model 생성
knn = KNeighborsClassifier(n_neighbors = 3)
#데이터셋 분배
X_train, X_test, Y_train, Y_test = train_test_split(iris['data'], iris['target'], test_size=0.25, stratify = iris['target'])
#학습
knn.fit(X_train, Y_train)
#예측
k_pred = knn.predict(X_test)
#정확도 출력
print("prediction accuracy: {:.2f}".format(np.mean(k_pred == Y_test)))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

#2. LogisticRegression
#전처리
scaler = StandardScaler()
X_train_t = scaler.fit_transform(X_train)
X_test_t = scaler.fit_transform(X_test)
#logisticregression model 생성
logistic_model = LogisticRegression()
#학습
logistic_model.fit(X_train_t, Y_train)
#예측
l_pred = logistic_model.predict(X_test_t)
#정확도 출력
print(logistic_model.score(X_test_t, Y_test))
```

```python
#3. SVM
from sklearn.svm import SVC

#전처리
sc = StandardScaler()
sc.fit(X_train)
X_train_t = sc.transform(X_train)
X_test_t = sc.transform(X_test)
#SVM model 생성
svm_model = SVC(kernel='rbf', C=8, gamma=0.1)
#학습
svm_model.fit(X_train_t, Y_train)
#예측
s_pred = svm_model.predict(X_test_t)
#출력
print("prediction accuracy: {:.2f}".format(np.mean(s_pred == Y_test)))
```

```python
#4. 결정트리

from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#decisiontree 생성
dt_clf = DecisionTreeClassifier(random_state=11)
#학습
model_all_params = dt_clf.fit(X_train, Y_train)
#예측
pred = dt_clf.predict(X_test)
#정확도 출력
print(accuracy_score(Y_test,pred))
```

```python
#5. AdaBoosting
from sklearn.ensemble import AdaBoostClassifier

#adaboosting 모델 생성 & 예측 & 정확도 출력
AdaBoost = AdaBoostClassifier(n_estimators=50, learning_rate=0.01)
model = AdaBoost.fit(X_train, Y_train)
print(model.score(X_test, Y_test))
```
